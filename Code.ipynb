```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


# Import the dataset into the code
df = pd.read_csv('sleep_health_and_lifestyle_dataset.csv')

df = df.drop('Person ID', axis=1)

df.loc[df['BMI Category'] == 'Normal Weight', 'BMI Category'] = 'Normal'

# Compute the statistical summary of the dataset using pandas describe()
# include='all' will include non-numeric columns of the dataset
summary = df.describe(include='all')

fig = plt.figure(figsize=(18, 14))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

plt.subplot(2, 3, 1)
# Visualization 1: Gender Distribution
# Plotting the bar plot for gender distribution
sns.set(style="whitegrid")
gender_distribution = df['Gender'].value_counts()
sns.barplot(x=gender_distribution.index,
            y=gender_distribution.values, palette="viridis")

plt.title('Gender Distribution in the Dataset')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.subplot(2, 3, 2)
# Visualization 2: Box plot for Sleep Duration
sns.boxplot(x='Sleep Duration', data=df, color='skyblue')
plt.title('Box Plot - Sleep Duration')

plt.subplot(2, 3, 3)
# Visualization 3: Histogram for Age
sns.histplot(df['Age'], bins=20, kde=True, color='salmon')
plt.title('Histogram - Age Distribution')
plt.xlabel('Age')

plt.subplot(2, 3, 4)
# Visualization 4: Bar plot for BMI Category
sns.countplot(x='BMI Category', data=df, palette='viridis')
plt.title('Bar Plot - BMI Category Distribution')
plt.xlabel('BMI Category')
plt.ylabel('Count')

# Summary text
summary = '''
Student Name: Abdul Rauf
Student ID: 22094627

The gender distribution bar plot represents that there 182 samples for Male population and 178 sample 
for Female population were taken for this dataset. This distribution insures that we get equal results 
for irrespective of the gender of the subject.

The histogram plot of the age distribution shows the age range of our samples and we can see that the 
age range is from 26 to 60. With age 64 having the largest sample size of 67. This distribution insures 
that we get the results for the adult population and which age group is affected the most.

The Box plot show the number of hours sleep range of our population with the minimum number of hours 
being 6.4 and the maximum number of hours being 7.8. This variation shows that we get the results for 
the all the sleep ranges that people sleep on an average.

Lastly, the Bar plot of the BMI distribution shows that most of the people have Normal weight category 
with number of entries being 230. The overweight category people also has a high number of entries as 
148. And the least amount of people are Obese with only 17 people affected by obesity in our dataset.

'''
# Title text
title = '''The average sleep pattern and BMI category of adult population around the world'''

# Add text to figure
plt.figtext(0.35, 0.1, summary, fontsize=15)
plt.figtext(0.15, 0.48, title, fontsize=24)

plt.show()
```
